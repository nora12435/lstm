```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

x_values = [1, 2, 3]

wf = 0.5
whf = 0.1
bf = 0

wi = 0.6
whi = 0.2
bi = 0

wc = 0.7
whc = 0.3
bc = 0

wo = 0.8
who = 0.4
bo = 0

wy = 4
by = 0

h = 0
c = 0

print("LSTM numerical example")
print()

for step in range(len(x_values)):
    x = x_values[step]

    f = sigmoid(wf * x + whf * h + bf)

    i = sigmoid(wi * x + whi * h + bi)

    c_bar = math.tanh(wc * x + whc * h + bc)

    c = f * c + i * c_bar

    o = sigmoid(wo * x + who * h + bo)

    h = o * math.tanh(c)

    print("step", step + 1)
    print("x =", x)
    print("f =", round(f, 3))
    print("i =", round(i, 3))
    print("c_bar =", round(c_bar, 3))
    print("c =", round(c, 3))
    print("o =", round(o, 3))
    print("h =", round(h, 3))
    print()

x = 4

f = sigmoid(wf * x + whf * h + bf)
i = sigmoid(wi * x + whi * h + bi)
c_bar = math.tanh(wc * x + whc * h + bc)
c = f * c + i * c_bar
o = sigmoid(wo * x + who * h + bo)
h = o * math.tanh(c)

y = wy * h + by

print("prediction step")
print("h4 =", round(h, 3))
print("predicted value =", round(y, 3))
```
